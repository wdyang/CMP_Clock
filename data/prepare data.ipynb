```python
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
# Load data
timeline = pd.read_csv('timeline.csv', sep='\t')
co2_26 = pd.read_csv('CleanedData/CO2Emissions/CO2EmissionsWithRCP2.6.csv')
co2_85 = pd.read_csv('CleanedData/CO2Emissions/CO2EmissionsWithRCP8.5.csv')
land_26 = pd.read_csv('CleanedData/LandUse/LandUseWithRCP2.6.csv')
land_85 = pd.read_csv('CleanedData/LandUse/LandUseWithRCP8.5.csv')
population_26 = pd.read_csv('CleanedData/Population/PopulationWithRCP2.6.csv')
population_85 = pd.read_csv('CleanedData/Population/PopulationWithRCP8.5.csv')
```

```python
# per data type aggregation
base = population_26['population'][0]
population_26['p_change_26'] = population_26['population']/base

base = population_85['population'][0]
population_85['p_change_85'] = population_85['population']/base

total_land= (land_26['land_area']+land_26['natural_land'])[0]
land_26['land_percent_26'] = land_26['land_area'] / total_land
land_85['land_percent_85'] = land_85['land_area'] / total_land

base = co2_26['CO2Emissions'][0]
co2_26['co2_change_26'] = co2_26['CO2Emissions'] / base

base = co2_85['CO2Emissions'][0]
co2_85['co2_change_85'] = co2_85['CO2Emissions'] / base

land_26 = land_26.rename(columns = {'land_area': 'land_26'})
land_85 = land_85.rename(columns = {'land_area': 'land_85'})
population_26 = population_26.rename(columns = {'population': 'population_26'})
population_85 = population_85.rename(columns = {'population': 'population_85'})
co2_26 = co2_26.rename(columns = {'CO2Emissions':'co2_26'})
co2_85 = co2_85.rename(columns = {'CO2Emissions':'co2_85'})
```

```python
# Merge into single data file
cols = ['Year', 'Seconds', 'RCP', 'land_26', 'land_percent_26']
df0=timeline.merge(land_26, left_on='Year', right_on = 'year', how='left')[cols]

cols.append('land_85'); cols.append('land_percent_85')
df1 = df0.merge(land_85, left_on='Year', right_on = 'year', how='left')[cols]

cols.append('population_26'); cols.append('p_change_26')
df1 = df1.merge(population_26, left_on='Year', right_on = 'year', how='left')[cols]

cols.append('population_85'); cols.append('p_change_85')
df1 = df1.merge(population_85, left_on='Year', right_on = 'year', how='left')[cols]

cols.append('co2_26')
cols.append('co2_change_26')
df1 = df1.merge(co2_26, left_on='Year', right_on = 'year', how='left')[cols]

cols.append('co2_85')
cols.append('co2_change_85')
df1 = df1.merge(co2_85, left_on='Year', right_on = 'year', how='left')[cols]
```
